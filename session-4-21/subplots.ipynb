```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
```

# Subplots

First recall what we did before. Below is code to make the same example array and plot it using pcolormesh:

```python
#make sample array
sample_array = np.arange(-50*9,50*9,9)
sample_array = np.tan(sample_array) # this is the tangent function from trig, if you're curious
sample_array = sample_array.reshape((10,10))

#make colormap
top = cm.get_cmap('Blues_r')
bottom = cm.get_cmap('Reds')
newcolors = np.vstack((top(np.linspace(0, 1, 10)),
                       bottom(np.linspace(0, 1, 10))))
newcmp = ListedColormap(newcolors, name='RedBlue')

#plot array with colorbar:
plt.pcolormesh(sample_array,cmap=newcmp)
plt.colorbar()
```

---------------------

To have multiple plots on the same image we make subplots. You may have made subplots before. To specify how many subplots we want in a plot:

``fig, ax = plt.subplots(nrow,ncolumns)`` , where nrows is the number of rows and ncolumns is the number of columns. Here is an example to make it more clear:

```python
fig, ax = plt.subplots(2,3)
#fig is how we refer to the whole figure of subplots
```

^^ As you can see, this makes a plot with 2 rows of subplots and 3 columns of subplots. 


To plot actual things in the subplot areas, we do ``ax[row, column].plot()``. In our case we would use ``ax[row, column].pcolormesh()``because we are plotting colormaps.

For example to plot something in the second row in the third column we would do this:

```python
fig, ax = plt.subplots(2,3)
ax[1,2].pcolormesh(sample_array,cmap=newcmp)
```
